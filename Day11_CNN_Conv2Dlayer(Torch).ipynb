```python
import torch
import torch.nn as nn
# Creating a Conv2D layer (same as Keras Conv2D)
conv = nn.Conv2d(
    in_channels = 1,     # grayscale input image
    out_channels = 32,   # 32 filters -> 32 feature maps
    kernel_size = 3,     # 3x3 filter
    stride = 1,          # filter moves 1 pixel at a time
    padding = 1          # same padding -> output size stays same
)
# ReLU activation function
relu = nn.ReLU()
# Creating a random input image
# Format: (batch_size, channels, height, width)
x = torch.randn(1, 1, 28, 28)
# Passing image through convolution layer
out = conv(x)
# Applying ReLU activation
out = relu(out)
# Print output shape
print("Output Shape:", out.shape)
# Inspect learnable parameters
print("Weights:", conv.weight.shape)
print("Bias   :", conv.bias.shape)
# Calculate total parameters
total_params = 32 * 1 * 3 * 3 + 32
print("Total:", total_params)
```
