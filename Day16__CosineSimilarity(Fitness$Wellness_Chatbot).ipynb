# Scenario: Fitness & Wellness Chatbot Training
You feed it these sentences:
- “Push-ups strengthen chest and triceps”
- “Yoga improves flexibility and balance”
- “Apples are rich in fiber”
- “Cardio exercises boost heart health”
👉 In this case, the chatbot learns to distinguish between exercise facts, nutrition facts, and general wellness statements, so it can tailor its responses — e.g., suggesting workouts, offering diet tips, or explaining health benefits.
Would you like me to design a third scenario (maybe in education or travel) so you have a set of examples across different industries?

```python
# Install library if not already installed
# !pip install sentence-transformers

from sentence_transformers import SentenceTransformer
import numpy as np

# Load the embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')


# Training sentences (knowledge base)
sentences = [
    "Push-ups strengthen chest and triceps",
    "Yoga improves flexibility and balance",
    "Apples are rich in fiber",
    "Cardio exercises boost heart health"
]


# Categories corresponding to sentences
categories = [
    "Exercise",
    "Exercise",
    "Nutrition",
    "Wellness"
]


# Convert training sentences to embeddings
sentence_embeddings = model.encode(sentences)


# Cosine similarity function
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))


# User query
query = "What workout helps strengthen my chest?"


# Convert query into embedding
query_embedding = model.encode(query)


# Calculate similarity between query and training sentences
similarities = []

for emb in sentence_embeddings:
    sim = cosine_similarity(query_embedding, emb)
    similarities.append(sim)


# Find the best matching sentence
best_match_index = np.argmax(similarities)


# Output results
print("User Query:", query)
print("Most Similar Sentence:", sentences[best_match_index])
print("Category:", categories[best_match_index])
print("Similarity Score:", similarities[best_match_index])
```
