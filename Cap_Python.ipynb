WAP TO FIND N amstrong number using function

```python
def is_armstrong(num):
    digits = len(str(num))
    temp = num
    total = 0

    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10

    return total == num


n = int(input("Enter value of n: "))
count = 0
num = 1

print(f"First {n} Armstrong numbers are:")

while count < n:
    if is_armstrong(num):
        print(num, end=" ")
        count += 1
    num += 1
```

WAP to find nth prime number



```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

def find_nth_prime(n):
    count = 0
    num = 2
    while count < n:
        if is_prime(num):
            count += 1
            if count == n:
                return num
        num += 1
find_nth_prime(615)
```

Create an abstract class shape with an abstract method area and  then define two subclasses as rectangle and circle that inherit from shape  circle  and implement the area method to find the area of rectangle and circle.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    def area(self):
        return self.length * self.breadth

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius
rect = Rectangle(10, 5)
cir = Circle(7)


print("Area of Rectangle:", rect.area())
print("Area of Circle:", cir.area())
```

```python
# Single Level
class Parent:
  a = 10
  b = 20

  def __init__(self , p , q):
    self.p = p
    self.q = q
  def disp(self):
    print(self.p , self.q)

class Child(Parent):
  x = 50
  y = 60
  def __init__(self,m,n,p,q):
    super().__init__(p,q)
    self.m = m
    self.n = n
  def show(self):
    super().disp()
    print(self.m , self.n)

ob = Child(100,200,300,400)
ob.show()
```

WAP to create a generator funtion to generate factorial number between the range 1- 100 and store the output in t he form of tuple

```python
def Fact(num):
  for i in range(1,num+1):
    fact = 1
    for j in range(1,i+1):
      fact = fact*j
    yield fact

print(tuple(Fact(100)))


```


WAP to find the square of all integers present in a range of 1 - n only if it is palindrome.

```python
def Palindrome(num):
  temp = num
  rev = 0
  while temp > 0:
    digit = temp%10
    rev = rev*10 + digit
    temp = temp//10
  if num == rev:
    return True
  else:
    return False

def Square(num):
  for i in range(1,num+1):
    if Palindrome(i):
      yield i**2
print(tuple(Square(1000)))
```

Create a decorator to calculate the execution time of the program.

```python
import time
def execution(func):
  def inner(*args,**kwargs):
    start = time.time()
    func(*args,**kwargs)
    end = time.time()
    print("Executin time is : " , end-start)
  return inner

@execution
def sqr():
  for i in range(1,10001):
    print(i)
sqr()
```

```python
from flask import Flask, jsonify, request
from pyngrok import ngrok

app = Flask(__name__)

@app.route("/")
def home():
    return jsonify({"message": "Flask API running"})

@app.route("/user", methods=["GET", "POST"])
def users():
    if request.method == "GET":
        return jsonify({"users": ["Alice", "Bob"]})
    if request.method == "POST":
        data = request.json
        return jsonify({"created user": data}), 201

# Start ngrok tunnel
public_url = ngrok.connect(5000)
print("Public URL:", public_url)

app.run(port=5000)
```
