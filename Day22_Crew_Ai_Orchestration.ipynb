```python
# Scenario: AI Content Automation for a Startup
# 🏢 Problem Statement

# A startup wants to:

# Stay updated with latest AI trends

# Automatically create LinkedIn posts

# Save time on manual research & writing

# 👉 So we build a multi-agent AI system to automate this using CrewAI.

!pip install crewai litellm langchain-community langchain-litellm

import os
os.environ["CREWAI_TRACING_ENABLED"] = "true"
from crewai import Agent, Task, Crew
from google.colab import userdata
from langchain_litellm import ChatLiteLLM # Updated import

# -----------------------------
# 1. Set Groq API Key
# -----------------------------
os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")

# ❌ REMOVE any OPENAI key completely
if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]

# -----------------------------
# 2. Define LLM (using ChatLiteLLM for Groq)
# -----------------------------
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# -----------------------------
# 3. Define Agents
# -----------------------------
researcher = Agent(
    role="Researcher",
    goal="Find latest AI news",
    backstory="Expert in AI trends",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

writer = Agent(
    role="Writer",
    goal="Write LinkedIn post",
    backstory="Expert content creator",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

# -----------------------------
# 4. Tasks
# -----------------------------
research_task = Task(
    description="Find 5 latest AI news trends",
    expected_output="List with summaries",
    agent=researcher
)

write_task = Task(
    description="Write LinkedIn post",
    expected_output="Final post",
    agent=writer,
    context=[research_task]
)

# -----------------------------
# 5. Crew
# -----------------------------
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
    verbose=True
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n✅ FINAL OUTPUT:\n")
print(result)
```

```python
# Scenario: Smart Job Application Assistant
# 🏢 Problem Statement

# Students struggle to:

# Analyze job descriptions

# Match their skills

# Write tailored resumes & cover letters

# 👉 So we build a multi-agent AI system that automates job application preparation.

# 🤖 System Overview
# 🔹 Agent 1: Job Analyzer

# Reads job description

# Extracts required skills

# 🔹 Agent 2: Resume Optimizer

# Matches candidate skills with job

# Suggests improvements

# 🔹 Agent 3: Cover Letter Writer

# Generates a personalized cover letter

# 🔄 Workflow
# Job Description → Skill Analysis → Resume Optimization → Cover Letter Generation


!pip install crewai litellm

import os
from crewai import Agent, Task, Crew
from google.colab import userdata

# -----------------------------
# 1. API Key Setup
# -----------------------------
os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")

# -----------------------------
# 2. Define Agents
# -----------------------------
job_analyzer = Agent(
    role="Job Analyzer",
    goal="Extract key skills and requirements from job descriptions",
    backstory=(
        "You are an HR expert who understands job postings and identifies "
        "important skills, qualifications, and expectations."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

resume_optimizer = Agent(
    role="Resume Optimizer",
    goal="Align candidate resume with job requirements",
    backstory=(
        "You are a career coach who improves resumes by matching them "
        "with job descriptions and highlighting relevant skills."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

cover_letter_writer = Agent(
    role="Cover Letter Writer",
    goal="Write a personalized cover letter",
    backstory=(
        "You are a professional writer who crafts compelling cover letters "
        "tailored to specific job roles."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

# -----------------------------
# 3. Input Data (Scenario)
# -----------------------------
job_description = """
We are looking for a Data Analyst with experience in Python, SQL, Power BI,
and data visualization. The candidate should be able to analyze datasets,
create dashboards, and communicate insights effectively.
"""

candidate_resume = """
I am a BTech student skilled in Python and basic SQL.
I have created Power BI dashboards and completed projects on data analysis.
"""

# -----------------------------
# 4. Tasks
# -----------------------------
analysis_task = Task(
    description=f"""
    Analyze the following job description and extract:
    - Required skills
    - Key responsibilities
    - Tools/technologies

    Job Description:
    {job_description}
    """,
    expected_output="Structured list of skills, responsibilities, and tools",
    agent=job_analyzer
)

resume_task = Task(
    description=f"""
    Compare the candidate resume with the job requirements and:
    - Identify skill gaps
    - Suggest improvements
    - Optimize the resume

    Resume:
    {candidate_resume}
    """,
    expected_output="Improved resume with suggestions",
    agent=resume_optimizer,
    context=[analysis_task]
)

cover_letter_task = Task(
    description="""
    Create a personalized cover letter using:
    - Job requirements
    - Candidate resume
    - Improvements suggested

    Make it professional and engaging.
    """,
    expected_output="Final cover letter",
    agent=cover_letter_writer,
    context=[analysis_task, resume_task]
)

# -----------------------------
# 5. Crew Setup
# -----------------------------
crew = Crew(
    agents=[job_analyzer, resume_optimizer, cover_letter_writer],
    tasks=[analysis_task, resume_task, cover_letter_task],
    verbose=True,
    tracing=True   # good for demo
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n🚀 FINAL OUTPUT:\n")
print(result)
```

# 📌 Problem Statement
Enterprise sales teams often struggle to:
- Analyze client requirements from RFPs (Requests for Proposals)
- Match company offerings to client needs
- Draft tailored proposals that stand out
- Ensure consistency across documents
👉 So we build a multi-agent AI system that automates sales proposal preparation.

🤖 System Overview
- 🔹 Agent 1: RFP Analyzer
- Reads client’s request for proposal
- Extracts key requirements, budget constraints, and evaluation criteria
- 🔹 Agent 2: Solution Mapper
- Matches company products/services to client needs
- Suggests tailored value propositions and differentiators
- 🔹 Agent 3: Proposal Writer
- Generates a polished, client-specific proposal
- Ensures alignment with brand tone and compliance standards

🔄 Workflow
RFP Document → Requirement Analysis → Solution Mapping → Proposal Generation

```python
# ============================================
# 🧠 Multi-Agent Sales Proposal System
# ============================================

import re
from datetime import datetime

# ============================================
# 🔹 STATE OBJECT (Shared Memory)
# ============================================
state = {
    "rfp_text": "",
    "requirements": [],
    "budget": "",
    "timeline": "",
    "criteria": [],
    "solutions": [],
    "final_proposal": ""
}

# ============================================
# 🔹 AGENT 1: RFP ANALYZER
# ============================================
def rfp_analyzer(state):
    print("\n🧾 [RFP Analyzer] Working...")

    text = state["rfp_text"].lower()

    # Extract requirements (simple keyword-based)
    keywords = ["cloud", "security", "ai", "data", "migration", "analytics"]
    requirements = [word for word in keywords if word in text]

    # Extract budget (simple regex)
    budget_match = re.search(r'\$?\d+[kK]?', text)
    budget = budget_match.group() if budget_match else "Not specified"

    # Extract timeline
    if "month" in text:
        timeline = "Few months"
    elif "week" in text:
        timeline = "Few weeks"
    else:
        timeline = "Not specified"

    # Dummy criteria extraction
    criteria = ["cost", "scalability", "performance"]

    state["requirements"] = requirements
    state["budget"] = budget
    state["timeline"] = timeline
    state["criteria"] = criteria

    print("✅ Extracted Requirements:", requirements)
    return state


# ============================================
# 🔹 AGENT 2: SOLUTION MAPPER
# ============================================
def solution_mapper(state):
    print("\n🔧 [Solution Mapper] Working...")

    mapping_db = {
        "cloud": "AWS Cloud Services",
        "security": "Advanced Security Suite",
        "ai": "AI/ML Platform",
        "data": "Data Engineering Pipeline",
        "migration": "Cloud Migration Toolkit",
        "analytics": "Business Analytics Dashboard"
    }

    solutions = []

    for req in state["requirements"]:
        if req in mapping_db:
            solutions.append({
                "requirement": req,
                "solution": mapping_db[req],
                "value": "High performance, scalable, cost-effective"
            })

    state["solutions"] = solutions

    print("✅ Mapped Solutions:", solutions)
    return state


# ============================================
# 🔹 AGENT 3: PROPOSAL WRITER
# ============================================
def proposal_writer(state):
    print("\n✍️ [Proposal Writer] Working...")

    proposal = f"""
========================================
📄 SALES PROPOSAL
========================================

📅 Date: {datetime.now().strftime("%Y-%m-%d")}

🔹 Executive Summary:
We understand your requirements and propose a tailored solution to meet your needs.

🔹 Client Requirements:
{', '.join(state['requirements'])}

🔹 Proposed Solutions:
"""

    for sol in state["solutions"]:
        proposal += f"""
- Requirement: {sol['requirement']}
  Solution: {sol['solution']}
  Value: {sol['value']}
"""

    proposal += f"""
🔹 Budget: {state['budget']}
🔹 Timeline: {state['timeline']}

🔹 Why Choose Us?
- Proven expertise
- Scalable solutions
- Cost optimization

🔹 Evaluation Criteria Addressed:
{', '.join(state['criteria'])}

========================================
✅ End of Proposal
========================================
"""

    state["final_proposal"] = proposal

    print("✅ Proposal Generated!")
    return state


# ============================================
# 🔹 HUMAN REVIEW (OPTIONAL)
# ============================================
def human_review(state):
    print("\n👨‍💼 [Human Review] Checking proposal...")

    decision = input("Approve proposal? (yes/no): ").strip().lower()

    if decision == "yes":
        print("✅ Approved!")
    else:
        print("❌ Rejected! सुधार needed.")
        state["final_proposal"] = "Proposal needs revision."

    return state


# ============================================
# 🔄 WORKFLOW (PIPELINE)
# ============================================
def run_system():
    print("\n🚀 Enterprise Sales Multi-Agent System")
    print("Type 'exit' to quit\n")

    while True:
        user_input = input("📥 Enter RFP Text: ")

        if user_input.lower() == "exit":
            print("👋 Exiting system...")
            break

        # Reset state
        global state
        state = {
            "rfp_text": user_input,
            "requirements": [],
            "budget": "",
            "timeline": "",
            "criteria": [],
            "solutions": [],
            "final_proposal": ""
        }

        # 🔄 Agent Workflow
        state = rfp_analyzer(state)
        state = solution_mapper(state)
        state = proposal_writer(state)

        # Optional Human Review
        state = human_review(state)

        # 📄 Final Output
        print("\n📄 FINAL OUTPUT:")
        print(state["final_proposal"])


# ============================================
# ▶️ RUN
# ============================================
if __name__ == "__main__":
    run_system()
```
