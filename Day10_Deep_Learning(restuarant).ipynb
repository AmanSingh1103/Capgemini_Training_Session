# Scenario:
 We want to classify whether a restaurant order is “Large” (1) or “Small” (0) based on the number
  of items ordered.
 - If items ≥ 3 → Large order
 - If items < 3 → Small order

```python
import numpy as np

# Step 1: Data Preparation
# X = number of items ordered
# y = order type (0 = Small, 1 = Large)

X = np.array([[1], [2], [3], [4]])   # input: number of items
y = np.array([0, 0, 1, 1])           # output: small=0, large=1

# Step 2: Initialize weights and bias
weights = np.random.rand(1)   # single input -> single weight
bias = np.random.rand(1)

learning_rate = 0.1

# Step 3: Activation Function (Step Function)
def activation(z):
    return 1 if z >= 0 else 0

# Step 4: Training Loop
for epoch in range(10):  # run for 10 iterations
    print(f"\nEpoch {epoch+1}")

    for i in range(len(X)):

        # Linear Combination
        z = np.dot(X[i], weights) + bias

        # Prediction
        y_pred = activation(z)

        # Error
        error = y[i] - y_pred

        # Update rule
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Items: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}")

# Final learned parameters
print("\nFinal Weights:", weights)
print("Final Bias:", bias)

# Step 5: Testing the perceptron
test_orders = np.array([1, 2, 3, 5])

print("\nTesting New Orders:")
for items in test_orders:
    z = np.dot(items, weights) + bias
    prediction = activation(z)

    order_type = "Large Order" if prediction == 1 else "Small Order"

    print(f"Items Ordered: {items}, Prediction: {order_type}")
```
