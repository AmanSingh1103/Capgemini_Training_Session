# Scenario: Smart Shopping Assistant 🛒

You say to your students:
“Imagine you are planning a party and you tell your assistant:
‘Find me the best chocolate cake under ₹500.’
Now instead of just replying with text, your assistant actually thinks, checks different bakery websites, compares prices, and then gives you the final decision.”

```python
# Smart Shopping Assistant (Agentic AI Simulation)

# Step 1: Available cakes (data collected from "websites")
cakes = [
    {"name": "Chocolate Truffle", "price": 480, "rating": 4.5, "delivery_time": 30},
    {"name": "Dark Chocolate Cake", "price": 520, "rating": 4.7, "delivery_time": 25},
    {"name": "Chocolate Cream Cake", "price": 450, "rating": 4.2, "delivery_time": 40},
    {"name": "Choco Lava Cake", "price": 400, "rating": 4.0, "delivery_time": 35}
]
```

```python
# Step 2: User goal
budget = 500

print("🎯 Goal: Find best chocolate cake under ₹500\n")

# Step 3: Filter cakes within budget (Perception)
valid_cakes = []
for cake in cakes:
    if cake["price"] <= budget:
        valid_cakes.append(cake)
```

```python
# Step 4: Reasoning (compare cakes based on rating and delivery)
# Sort by highest rating, then fastest delivery
valid_cakes.sort(key=lambda x: (-x["rating"], x["delivery_time"]))

# Step 5: Decision
best_cake = valid_cakes[0]

# Step 6: Action (final output)
print("✅ Best Cake Found:")
print(f"Name: {best_cake['name']}")
print(f"Price: ₹{best_cake['price']}")
print(f"Rating: {best_cake['rating']} ⭐")
print(f"Delivery Time: {best_cake['delivery_time']} mins")
```

# Scenario: The University’s AI Plagiarism Agent
Imagine a university has deployed an AI-powered plagiarism detection agent named PlagiarismLLM.
Its job is to monitor student essay submissions in real time and decide whether they are original or potentially plagiarized.


Step 1: Thinking (Reasoning)
Whenever an essay is submitted, PlagiarismLLM first analyzes the text:
- Example: “Essay contains large verbatim passages from online sources”
- PlagiarismLLM thinks: “Analyzing essay: multiple verbatim passages detected.”

Step 2: Planning (Choosing an Action)
Based on the thought, PlagiarismLLM decides which tool to use:
- If the essay has long verbatim matches or suspicious similarity scores, it plans to flag as potential plagiarism.
- Otherwise, it plans to mark as original work.

Step 3: Acting (Running Tools)
PlagiarismLLM hands over the decision to PlagiarismTools, which executes the action:
- flag_plagiarism → returns “Potential Plagiarism Detected”
- mark_original → returns “Original Work”

Step 4: Answering (Final Decision)
Once a clear classification is reached, PlagiarismLLM announces the decision:
- “Decision: Potential Plagiarism Detected”
- “Decision: Original Work”


```python
# AI Plagiarism Agent Simulation

# -------------------------------
# Step 0: Sample Essays (input)
# -------------------------------
essays = [
    "This essay is completely original and written by the student.",
    "This essay contains copied content from Wikipedia and other online sources.",
    "The student has written most parts but some sentences are copied from internet."
]

# -------------------------------
# Step 1: Thinking (Reasoning)
# -------------------------------
def analyze_essay(essay):
    print("\n🧠 Thinking...")

    # Simple rule-based detection (simulation)
    suspicious_keywords = ["copied", "Wikipedia", "online sources", "internet"]

    score = 0
    for word in suspicious_keywords:
        if word.lower() in essay.lower():
            score += 1

    print(f"Analysis Score: {score}")
    return score

# -------------------------------
# Step 2: Planning (Decision)
# -------------------------------
def plan_action(score):
    print("📋 Planning...")

    if score >= 2:
        return "flag_plagiarism"
    else:
        return "mark_original"

# -------------------------------
# Step 3: Acting (Tool Execution)
# -------------------------------
def plagiarism_tools(action):
    print("🛠️ Acting...")

    if action == "flag_plagiarism":
        return "Potential Plagiarism Detected"
    else:
        return "Original Work"

# -------------------------------
# Step 4: Final Answer
# -------------------------------
def plagiarism_agent(essay):
    print("\n📄 New Essay Submitted:")
    print(essay)

    # Think
    score = analyze_essay(essay)

    # Plan
    action = plan_action(score)

    # Act
    result = plagiarism_tools(action)

    # Answer
    print("🎯 Final Decision:", result)

# -------------------------------
# Run Agent on all essays
# -------------------------------
for e in essays:
    plagiarism_agent(e)
```
