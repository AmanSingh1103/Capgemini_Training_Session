**Building MINI AI Tool**

```python
def ai_summary_tool(text):

    prompt = f"""
    Analyze the following article and provide:

    1. Short Summary
    2. Key Insights (3 bullet points)
    3. One Actionable Recommendation

    Article:
    {text}
    """

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role":"user","content":prompt}],
        temperature=0.6,
        max_tokens=300
    )

    return response.choices[0].message.content
input_article = """
  AI-powered healthcare systems can detect diseases earlier and
  support doctors with faster diagnosis. Hospitals are adopting
  machine learning tools for medical imaging and patient monitoring.
  """

result = ai_summary_tool(input_article)

print("AI TOOL OUTPUT:\n")
print(result)
```
