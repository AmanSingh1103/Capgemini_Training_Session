```python
import tensorflow as tf
from tensorflow import keras

# Define the Conv2D layer
conv_layer = keras.layers.Conv2D(
    # Number of filters (kernels) the layer will learn
    # Each filter detects a different pattern in the image
    # Example: edges, curves, textures etc.
    # 32 filters means the output will have 32 feature maps
    filters = 32,

    # Size of the filter (kernel)
    # (3,3) means the filter scans a 3x3 region of pixels at a time
    # This is the most common kernel size in CNNs
    kernel_size = (3, 3),

    # How much the filter moves each step
    # (1,1) means it moves 1 pixel at a time horizontally and vertically
    # Smaller stride = more detailed feature extraction
    strides = (1, 1),

    # Padding determines what happens at the image borders
    # 'same' means output size stays the SAME as input size
    # TensorFlow adds extra pixels (zeros) around the image if needed
    padding = 'same',

    # Activation function applied after convolution
    # ReLU = max(0, x)
    # It removes negative values and keeps positive ones
    # This helps neural networks learn complex patterns
    activation = 'relu',

    # Shape of the input image
    # (height, width, channels)
    # 28 = image height
    # 28 = image width
    # 1 = grayscale channel
    # Example: MNIST handwritten digit dataset
    input_shape = (28, 28, 1)
)

# Create a Sequential model with the defined Conv2D layer
model = keras.Sequential([
    conv_layer
])

# Show a summary of the model architecture
# This prints:
# - layer type
# - output shape
# - number of parameters
model.summary()
```
