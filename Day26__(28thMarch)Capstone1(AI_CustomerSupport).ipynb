# **Scenario 1: AI Customer Support Multi-Agent System**
🎯 Objective:
Build a multi-agent customer support system that can classify, delegate, and resolve user queries.

💼 Problem Statement
A company wants to automate its support system using AI agents.

The system should:

Understand customer queries
Route them to the correct department
Provide responses
👥 Agents to Build
Classifier Agent
Identifies query type (Billing / Technical / General)
Billing Agent
Handles payment/refund queries
Technical Support Agent
Handles app/software issues
Response Agent
Combines outputs into a final response
⚙️ Tasks to Implement
Multi-agent architecture (at least 3 agents mandatory)
Task delegation logic
Basic agent communication (message passing / function calls)
💡 Sample Inputs
“I was charged twice for my subscription”
“The app crashes when I open it”
“What are your pricing plans?”
🧪 Expected Output
Correct classification of query
Delegation to appropriate agent
Final structured response

```python
!pip install Groq
import os
from groq import Groq

# -----------------------------
# 🔑 Step 1: Take API Key Input
# -----------------------------
api_key = input("Enter your GROQ API Key: ")
client = Groq(api_key=api_key)

# -----------------------------
# 🤖 LLM CALL FUNCTION
# -----------------------------
def call_llm(prompt):
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )
    return response.choices[0].message.content.strip()


# -----------------------------
# 🧠 CLASSIFIER AGENT
# -----------------------------
def classifier_agent(query):
    prompt = f"""
    Classify the following customer query into one category:
    Billing, Technical, or General.

    Query: "{query}"

    Only return one word.
    """
    category = call_llm(prompt)
    return category.lower()


# -----------------------------
# 💳 BILLING AGENT
# -----------------------------
def billing_agent(query):
    prompt = f"""
    You are a billing support agent.
    Handle the following query professionally:

    Query: "{query}"
    """
    return call_llm(prompt)


# -----------------------------
# 🛠️ TECHNICAL AGENT
# -----------------------------
def technical_agent(query):
    prompt = f"""
    You are a technical support agent.
    Help solve the issue:

    Query: "{query}"
    """
    return call_llm(prompt)


# -----------------------------
# 📘 GENERAL AGENT
# -----------------------------
def general_agent(query):
    prompt = f"""
    You are a customer support assistant.
    Answer general questions clearly:

    Query: "{query}"
    """
    return call_llm(prompt)


# -----------------------------
# 🧩 RESPONSE AGENT
# -----------------------------
def response_agent(category, response):
    final_output = f"""
    🔹 Query Type: {category.upper()}

    🔹 Response:
    {response}
    """
    return final_output


# -----------------------------
# 🔄 MAIN ORCHESTRATION
# -----------------------------
def process_query(query):
    print("\n[Classifier Agent] Processing...")
    category = classifier_agent(query)

    print(f"[Routing] Query classified as: {category}")

    if "billing" in category:
        response = billing_agent(query)

    elif "technical" in category:
        response = technical_agent(query)

    else:
        response = general_agent(query)

    final = response_agent(category, response)
    return final


# -----------------------------
# 🧪 TEST RUN
# -----------------------------
if __name__ == "__main__":
    print("\n=== AI Customer Support System ===\n")

    while True:
        user_query = input("\nEnter your query (or type 'exit'): ")

        if user_query.lower() == "exit":
            break

        result = process_query(user_query)
        print("\n✅ FINAL RESPONSE:")
        print(result)
```
