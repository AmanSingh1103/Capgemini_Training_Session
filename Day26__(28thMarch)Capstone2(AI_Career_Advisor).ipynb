# **🔷 Scenario 2: AI Career Advisor & Learning Path Generator**
🎯 Objective:
Build a multi-step intelligent advisor system using agent workflows.

💼 Problem Statement
Students need guidance on career paths based on their skills.

Your system should:

Analyze student profile
Suggest career roles
Generate learning roadmap
👥 Agents to Build
Profile Analyzer Agent
Understands student background
Career Recommendation Agent
Suggests suitable roles
Skill Gap Agent
Identifies missing skills
Learning Path Agent
Creates step-by-step roadmap
⚙️ Tasks to Implement
Sequential workflow (LangGraph preferred)
State passing between agents
Conditional logic (beginner vs intermediate)
💡 Sample Input
“I know Python basics and statistics, and I want to become a Data Scientist”

🧪 Expected Output
Suggested roles (e.g., Data Scientist, Analyst)
Skill gaps (ML, SQL, etc.)
Learning roadmap (step-by-step)
🧰 Tools (Choose Any)
LangGraph (recommended)
CrewAI
AutoGen
Python logic-based workflow
Microsoft Copilot Studio
📦 Deliverables
Workflow implementation
Output for at least 2 student profiles
Diagram of agent flow

```python
!pip install groq
import os
from groq import Groq

# -----------------------------
# 🔑 API KEY INPUT
# -----------------------------
api_key = input("Enter your GROQ API Key: ")
client = Groq(api_key=api_key)

MODEL = "llama-3.1-8b-instant"

# -----------------------------
# 🤖 LLM CALL
# -----------------------------
def call_llm(prompt):
    response = client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )
    return response.choices[0].message.content.strip()


# -----------------------------
# 🧠 AGENT 1: PROFILE ANALYZER
# -----------------------------
def profile_analyzer(state):
    query = state["input"]

    prompt = f"""
    Analyze the student profile and determine:
    1. Skills
    2. Experience level (beginner/intermediate)

    Input: "{query}"

    Output format:
    Skills: ...
    Level: ...
    """

    result = call_llm(prompt)

    state["profile"] = result

    # Extract level (simple logic)
    if "beginner" in result.lower():
        state["level"] = "beginner"
    else:
        state["level"] = "intermediate"

    return state


# -----------------------------
# 🎯 AGENT 2: CAREER RECOMMENDER
# -----------------------------
def career_agent(state):
    profile = state["profile"]

    prompt = f"""
    Based on this profile, suggest 2-3 suitable career roles.

    Profile:
    {profile}
    """

    roles = call_llm(prompt)
    state["career_roles"] = roles
    return state


# -----------------------------
# 🧩 AGENT 3: SKILL GAP ANALYZER
# -----------------------------
def skill_gap_agent(state):
    roles = state["career_roles"]
    level = state["level"]

    prompt = f"""
    Identify missing skills required for these roles.

    Roles:
    {roles}

    Current Level: {level}

    List skill gaps clearly.
    """

    gaps = call_llm(prompt)
    state["skill_gaps"] = gaps
    return state


# -----------------------------
# 📘 AGENT 4: LEARNING PATH
# -----------------------------
def learning_path_agent(state):
    gaps = state["skill_gaps"]
    level = state["level"]

    if level == "beginner":
        difficulty = "start from basics step-by-step"
    else:
        difficulty = "focus on advanced and project-based learning"

    prompt = f"""
    Create a step-by-step learning roadmap.

    Skill gaps:
    {gaps}

    Instruction: {difficulty}

    Provide numbered steps.
    """

    roadmap = call_llm(prompt)
    state["learning_path"] = roadmap
    return state


# -----------------------------
# 🔄 WORKFLOW EXECUTION
# -----------------------------
def run_workflow(user_input):
    state = {
        "input": user_input,
        "profile": "",
        "career_roles": "",
        "skill_gaps": "",
        "learning_path": "",
        "level": ""
    }

    print("\n[Agent 1] Profile Analyzer...")
    state = profile_analyzer(state)

    print("[Agent 2] Career Recommendation...")
    state = career_agent(state)

    print("[Agent 3] Skill Gap Analysis...")
    state = skill_gap_agent(state)

    print("[Agent 4] Learning Path Generation...")
    state = learning_path_agent(state)

    return state


# -----------------------------
# 🧪 MAIN
# -----------------------------
if __name__ == "__main__":
    print("\n=== AI Career Advisor System ===\n")

    while True:
        user_input = input("\nEnter student profile (or 'exit'): ")

        if user_input.lower() == "exit":
            break

        result = run_workflow(user_input)

        print("\n✅ FINAL OUTPUT:\n")
        print("🔹 Profile Analysis:\n", result["profile"])
        print("\n🔹 Career Roles:\n", result["career_roles"])
        print("\n🔹 Skill Gaps:\n", result["skill_gaps"])
        print("\n🔹 Learning Roadmap:\n", result["learning_path"])
```
