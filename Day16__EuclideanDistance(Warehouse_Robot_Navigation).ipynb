 # Scenario: Warehouse Robot Navigation
Imagine a warehouse where a robot needs to pick up items.
- The robot is currently at position (1,2) on the warehouse grid.
- The item it needs to collect is stored at position (4,6).
The robot’s navigation system calculates the straight-line distance to the item so it knows how far it must travel if it moves directly.

```python
# Simple Euclidean Distance example for Warehouse Robot Navigation
import numpy as np

# Robot current position (x, y)
robot_position = np.array([1, 2])

# Item position in warehouse (x, y)
item_position = np.array([4, 6])

# Calculate Euclidean Distance
distance = np.linalg.norm(robot_position - item_position)

# Print results
print("Robot Position:", robot_position)
print("Item Position:", item_position)
print("Distance to Item:", distance)

# Interpretation
print("The robot must travel approximately", distance, "units to reach the item.")
```
