# **Agriculture_yield**

```python




import numpy as np

# Crop yields (tons per hectare) for 3 crops across 4 regions
yields = np.array([[2.5, 3.0, 1.8],
                   [2.8, 3.2, 2.0],
                   [3.1, 2.9, 2.2],
                   [2.6, 3.1, 1.9]])

print("=== Agriculture Yield (Broadcasting + Indexing) ===\n")
print("Original crop yields (tons per hectare):")
print("         Crop A  Crop B  Crop C")
print(yields)
print()

# 1. Increase yields of the second region by [0.2, 0.3, 0.1]
adjustment = np.array([0.2, 0.3, 0.1])
print("1. Increase yields of second region (index 1) by [0.2, 0.3, 0.1]:")
print(f"   Original region 1: {yields[1]}")
yields[1] = yields[1] + adjustment  # Broadcasting + row indexing
print(f"   Adjusted region 1: {yields[1]}")
print()
print("   Updated yields:")
print("         Crop A  Crop B  Crop C")
print(yields)
print()

# 2. Multiply the last column (crop C) by 1.1 to simulate fertilizer impact
print("2. Multiply last column (Crop C) by 1.1 for fertilizer impact:")
print(f"   Original Crop C yields: {yields[:, 2]}")
yields[:, 2] = yields[:, 2] * 1.1  # Column indexing
print(f"   After fertilizer Crop C: {yields[:, 2]}")
print()
print("   Updated yields:")
print("         Crop A  Crop B  Crop C")
print(yields)
print()

# 3. Which crop shows the highest average yield across all regions?
avg_yields = np.mean(yields, axis=0)
crops = ['Crop A', 'Crop B', 'Crop C']
highest_crop_index = np.argmax(avg_yields)

print("3. Average yield across all regions:")
for i, crop in enumerate(crops):
    print(f"   {crop}: {avg_yields[i]:.2f} tons/hectare")
print()
print(f"   Highest average yield: {crops[highest_crop_index]} with {avg_yields[highest_crop_index]:.2f} tons/hectare")
```

# **Healthcare_dataset_index**

```python
import numpy as np

# Patient blood pressure readings [systolic, diastolic]
bp = np.array([[120, 80],
               [135, 85],
               [140, 90],
               [110, 70],
               [125, 75]])

print("=== Healthcare Dataset (Indexing) ===\n")
print("Original blood pressure readings [systolic, diastolic]:")
print(bp)
print()

# 1. Extract the systolic values (first column)
systolic = bp[:, 0]
print("1. Systolic values (first column):")
print(f"   {systolic}")
print()

# 2. Find patients with systolic > 130 using boolean indexing
high_systolic_mask = systolic > 130
high_systolic_patients = np.where(high_systolic_mask)[0]
print("2. Patients with systolic > 130:")
print(f"   Boolean mask: {high_systolic_mask}")
print(f"   Patient indices: {high_systolic_patients}")
print(f"   Patient readings:")
for idx in high_systolic_patients:
    print(f"      Patient {idx}: {bp[idx]}")
print()

# 3. Replace all diastolic values < 80 with 80 (minimum threshold)
bp_corrected = bp.copy()  # Create a copy to preserve original
diastolic_mask = bp_corrected[:, 1] < 80
bp_corrected[diastolic_mask, 1] = 80

print("3. Replace diastolic values < 80 with 80:")
print(f"   Diastolic values < 80 mask: {diastolic_mask}")
print(f"   Corrected blood pressure readings:")
print(bp_corrected)
print()
print("   Changes made:")
for idx in np.where(diastolic_mask)[0]:
    print(f"      Patient {idx}: {bp[idx]} → {bp_corrected[idx]}")
```

# **Weather_data_broadcasting**

```python
import numpy as np

# Temperature readings (in Celsius) for 5 cities over 3 days
temps = np.array([[25, 28, 30],
                  [22, 24, 26],
                  [30, 32, 33],
                  [27, 29, 31],
                  [20, 21, 23]])

print("=== Weather Data (Broadcasting) ===\n")
print("Original temperatures (Celsius):")
print(temps)
print(f"Shape: {temps.shape} (5 cities × 3 days)\n")

# 1. Convert all temperatures to Fahrenheit using broadcasting
# Formula: F = C * 9/5 + 32
temps_fahrenheit = temps * 9/5 + 32
print("1. Temperatures converted to Fahrenheit:")
print(temps_fahrenheit)
print()

# 2. Add a correction factor [1, -1, 0] to adjust readings for sensor bias
correction_factor = np.array([1, -1, 0])
print("2. Correction factor:", correction_factor)
print("   Shape:", correction_factor.shape)
print()

# Apply correction using broadcasting
temps_corrected = temps + correction_factor
print("   Corrected temperatures (Celsius):")
print(temps_corrected)
print()

print("   Broadcasting explanation:")
print("   - Original temps shape: (5, 3) - 5 cities, 3 days")
print("   - Correction factor shape: (3,) - 3 values for 3 days")
print("   - NumPy broadcasts the correction factor across all 5 cities")
print("   - Day 1: +1°C, Day 2: -1°C, Day 3: +0°C for all cities")
print("   - Result shape: (5, 3) - same as original")
```

# **Farm Crop yeild Analysis**

```python
# Scenario: Farm Crop Yield Analysis
# A farm tracks the yield (in kilograms) of three different crops — Wheat, Rice, and Corn — across four seasons (Spring, Summer, Autumn, Winter).
# The farm manager wants to analyze the data to make better decisions for next year.
# Tasks
# - Create a 2D array representing crop yields for 3 crops across 4 seasons.
# - Rows = crops
# - Columns = seasons
# - Increase all yields by 10% to account for improved irrigation.
# - Extract only Wheat’s yields (the first crop).
# - Find the average yield per crop across all seasons.
# - Identify yields greater than 3000 kg (to see which crops/seasons performed exceptionally well).

import numpy as np

# Step 1: Create 2D array
yields = np.array([
    [2800, 3200, 3000, 2900],   # Wheat
    [3500, 3700, 3400, 3600],   # Rice
    [2500, 2700, 2600, 2800]    # Corn
])

print("Original Crop Yields (kg):")
print(yields)

# Step 2: Increase by 10%
improved_yields = yields * 1.10
print("\nYields after 10% increase:")
print(improved_yields)

# Step 3: Extract Wheat
wheat_yields = improved_yields[0]
print("\nWheat Yields:")
print(wheat_yields)

# Step 4: Average per crop
average_per_crop = np.mean(improved_yields, axis=1)
print("\nAverage Yield per Crop:")
print(average_per_crop)

# Step 5: Yields > 3000
high_yields = improved_yields[improved_yields > 3000]
print("\nYields greater than 3000 kg:")
print(high_yields)
```

# **Factory Production Analysis**

```python
import numpy as np

# Factory Production Analysis
# Products: Gadgets, Tools, Machines
# Quarters: Q1, Q2, Q3, Q4

# Create 2D array: 3 products x 4 quarters
production = np.array([
    [120, 150, 180, 200],  # Gadgets
    [300, 320, 350, 400],  # Tools
    [450, 480, 520, 550]   # Machines
])

print("Original Production Data:")
print("Products: Gadgets, Tools, Machines")
print("Quarters: Q1, Q2, Q3, Q4")
print(production)
print()

# Increase all production by 5% for efficiency improvements
production_improved = production * 1.05
print("Production after 5% efficiency improvement:")
print(production_improved)
print()

# Extract Gadgets' production (first row)
gadgets_production = production[0]
print("Gadgets production across all quarters:")
print(gadgets_production)
print()

# Find average production per product across all quarters
avg_per_product = np.mean(production, axis=1)
print("Average production per product:")
print(f"Gadgets: {avg_per_product[0]:.2f}")
print(f"Tools: {avg_per_product[1]:.2f}")
print(f"Machines: {avg_per_product[2]:.2f}")
print()

# Identify production values greater than 500 units
exceptional_performance = production > 500
print("Exceptional performance (>500 units):")
print(exceptional_performance)
print("\nValues exceeding 500 units:")
print(production[exceptional_performance])
```

Scenario: Corporate Employee Performance Dashboard
A company tracks quarterly performance scores of employees across different departments.
The HR analytics team wants to analyze this dataset to identify top performers, departmental trends, and highlight exceptional scores.

Dataset
The dataset contains the following columns:
- Employee: Name of the employee
- Department: Department they belong to (Sales, HR, IT, Finance, etc.)
- Quarter: The quarter in which the performance score was recorded (Q1, Q2, Q3, Q4)
- Score: Performance score (numeric value)

Tasks
- Filtering (Core Corporate Skill)
- Find employees with Score > 90 (exceptional performers).
- Filter records for the Sales department only.
- Sorting (Power Moment)
- Sort employees by Score in descending order to rank performance.
- Sort by Department first, then Score to see departmental rankings.
- Grouping (Core Corporate Skill)
- Group by Department and calculate the average score per department.
- Group by Quarter and find the maximum score achieved in each quarter.

Business Context
- Filtering helps HR identify exceptional performers and focus on specific departments.
- Sorting creates clear performance rankings, useful for promotions or recognition programs.
- Grouping reveals departmental strengths and quarterly performance trends, guiding training and resource allocation.

```python

import pandas as pd
import numpy as np
import os
from google.colab import files

# ----------------------------
# STEP 1: Upload Files
# ----------------------------
print("Please upload Customers, Sales, and Support CSV files")
uploaded = files.upload()

print("\nUploaded Files:")
print(list(uploaded.keys()))


# STEP 2: Auto Detect File Names
file_list = list(uploaded.keys())

customers_file = [f for f in file_list if "customer" in f.lower()][0]
sales_file = [f for f in file_list if "sales" in f.lower()][0]
support_file = [f for f in file_list if "support" in f.lower()][0]

print("\nDetected Files:")
print("Customers File:", customers_file)
print("Sales File:", sales_file)
print("Support File:", support_file)


# STEP 3: Load Data

customers = pd.read_csv(customers_file)
sales = pd.read_csv(sales_file)
support = pd.read_csv(support_file)

print("\nShapes:")
print("Customers:", customers.shape)
print("Sales:", sales.shape)
print("Support:", support.shape)


# STEP 4: Basic Inspection

print("\nMissing Values:")
print("\nCustomers:\n", customers.isnull().sum())
print("\nSales:\n", sales.isnull().sum())
print("\nSupport:\n", support.isnull().sum())


# STEP 5: Data Cleaning


# Standardize column names
customers.columns = customers.columns.str.strip()
sales.columns = sales.columns.str.strip()
support.columns = support.columns.str.strip()

# Rename Customer ID column
customers.rename(columns={"Customer ID": "CustomerID"}, inplace=True)
sales.rename(columns={"Customer ID": "CustomerID"}, inplace=True)
support.rename(columns={"Customer ID": "CustomerID"}, inplace=True)

# Convert date columns if present
if "Signup Date" in customers.columns:
    customers["Signup Date"] = pd.to_datetime(customers["Signup Date"], errors='coerce')

if "Order Date" in sales.columns:
    sales["Order Date"] = pd.to_datetime(sales["Order Date"], errors='coerce')

# Fill missing Age
if "Age" in customers.columns:
    customers["Age"] = customers["Age"].fillna(customers["Age"].median())


# STEP 6: Feature Engineering

if "Price" in sales.columns and "Quantity" in sales.columns:
    sales["DiscountedPrice"] = sales["Price"] * 0.9
    sales["Revenue"] = sales["Quantity"] * sales["Price"]


# STEP 7: Filtering Examples

if "Order Date" in sales.columns:
    jan_2025_orders = sales[
        (sales["Order Date"].dt.year == 2025) &
        (sales["Order Date"].dt.month == 1)
    ]
    print("\nJanuary 2025 Orders:")
    print(jan_2025_orders)

print("\nFirst 5 Sales Records:")
print(sales.head())


# STEP 8: Aggregations


# Merge region into sales
if "Region" in customers.columns:
    sales_with_region = sales.merge(
        customers[["CustomerID", "Region"]],
        on="CustomerID",
        how="left"
    )

    avg_revenue_region = sales_with_region.groupby("Region")["Revenue"].mean()
    print("\nAverage Revenue by Region:")
    print(avg_revenue_region)

# Average resolution time by issue
if "Issue Type" in support.columns and "Resolution Time" in support.columns:
    avg_resolution_issue = support.groupby("Issue Type")["Resolution Time"].mean()
    print("\nAverage Resolution Time by Issue Type:")
    print(avg_resolution_issue)


# STEP 9: Customer Lifetime Value (CLV)

if "Revenue" in sales.columns:
    clv = sales.groupby("CustomerID")["Revenue"].sum().reset_index()
    clv.rename(columns={"Revenue": "CLV"}, inplace=True)
else:
    clv = pd.DataFrame()


# STEP 10: Avg Resolution Time per Customer

if "Resolution Time" in support.columns:
    avg_res_customer = support.groupby("CustomerID")["Resolution Time"].mean().reset_index()
    avg_res_customer.rename(columns={"Resolution Time": "AvgResolutionTime"}, inplace=True)
else:
    avg_res_customer = pd.DataFrame()


# STEP 11: Final Merge
merged = sales.merge(customers, on="CustomerID", how="left") \
              .merge(support, on="CustomerID", how="left")

if not clv.empty:
    merged = merged.merge(clv, on="CustomerID", how="left")

if not avg_res_customer.empty:
    merged = merged.merge(avg_res_customer, on="CustomerID", how="left")

# STEP 12: Export Final Data
merged.to_csv("Cleaned_Data.csv", index=False)

print("\n✅ Cleaned_Data.csv exported successfully!")

# Auto Download
files.download("Cleaned_Data.csv")
```
