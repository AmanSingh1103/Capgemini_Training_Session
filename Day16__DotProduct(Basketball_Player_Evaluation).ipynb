# Scenario: Basketball Player Evaluation
A coach is evaluating how well a player’s skills match the team’s current needs.
- The team’s requirement vector is a = [2,5,1].
- 2 = need for shooting ability
- 5 = need for defense
- 1 = need for passing
- The player’s skill vector is b = [3,4,2].
- 3 = player’s shooting ability
- 4 = defensive ability
- 2 = passing ability
The coach calculates the dot product:

```python

import numpy as np

# Team requirement vector
# [shooting, defense, passing]
a = np.array([2,5,1])

# Player skill vector
# [shooting ability, defensive ability, passing ability]
b = np.array([3,4,2])

# Calculate dot product
result = np.dot(a,b)

print("Team Requirement Vector:", a)
print("Player Skill Vector:", b)
print("Dot Product Score:", result)

# Interpretation
print("Higher score means the player's skills match the team's needs better.")
```
